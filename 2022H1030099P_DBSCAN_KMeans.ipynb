```python
kmeans_kwargs = {
    "init": "k-means++",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features_iris)
    sse.append(kmeans.inertia_)
```

```python
 plt.style.use("fivethirtyeight")
 plt.plot(range(1, 11), sse)
 plt.xticks(range(1, 11))
 plt.xlabel("Number of Clusters")
 plt.ylabel("SSE")
 plt.show()
```

```python
df_dbscan=df.iloc[0:1000,0:2]
```

```python
df_dbscan.count()
```

```python
df_dbscan.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features_dbscan = scaler.fit_transform(df_dbscan)
```

```python
scaled_features_dbscan
```

```python
from sklearn.cluster import DBSCAN
```

```python
clustering = DBSCAN(eps=0.3, min_samples=50).fit(scaled_features_dbscan)
```

```python
df_dbscan['Clusters'] = clustering.labels_ 
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_dbscan['x'], df_dbscan['y'],c=df_dbscan['Clusters'], cmap='Paired')
plt.title("Clusters determined by DBSCAN")
```

```python
clustering.labels_[clustering.labels_ == -1].size
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
model = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
model.fit(scaled_features_dbscan)
labels = model.labels_
```

```python
df_dbscan['agg_clusters']=labels
```

```python
df_dbscan.head()
```

```python
dendrogram = sch.dendrogram(sch.linkage(scaled_features_dbscan, method='single'))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_dbscan['x'], df_dbscan['y'],c=df_dbscan['agg_clusters'], cmap='Paired')
plt.title("Clusters determined by agglomerative clustering linkage single")
```
